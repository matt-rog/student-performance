# Exploratory Data Analysis

```python
from ucimlrepo import fetch_ucirepo
import matplotlib.pyplot as plt

# fetch dataset
student_performance = fetch_ucirepo(id=320)

# data (as pandas dataframes)
X = student_performance.data.features
y = student_performance.data.targets
```

```python
plt.scatter(y['G1'], y['G2'])
plt.xlabel('G1')
plt.ylabel('G2')
plt.title('Scatter Plot of G1 vs G2')
plt.show()
```

```python
colors = ["#FFD1B2", "#B2D1FF"]

integer_features = student_performance.variables[(student_performance.variables['type'] == 'Integer') & (student_performance.variables['role'] == 'Feature')]['name'].tolist()
fig, axes = plt.subplots(nrows=1, ncols=len(integer_features), figsize=(15, 5))

fig.suptitle('Distribution of Integer Features', fontsize=16)

for i, feature in enumerate(integer_features):
    color = colors[i % len(colors)]
    axes[i].boxplot(
        X[feature],
        patch_artist=True,
        boxprops=dict(facecolor=color, color='black'),
        capprops=dict(color='black'),
        whiskerprops=dict(color='black'),
        flierprops=dict(markeredgecolor='black', markerfacecolor='black'),
        medianprops=dict(color='black')
    )
    axes[i].set_title(feature)
    if i == 0:
        axes[i].set_ylabel("Value")

plt.tight_layout()
plt.show()
```

```python
category_features = student_performance.variables[(student_performance.variables['type'] == 'Categorical') & (student_performance.variables['role'] == 'Feature')]['name'].tolist()
fig, axes = plt.subplots(nrows=1, ncols=len(category_features), figsize=(15, 5))

fig.suptitle('Distribution of Categorical Features', fontsize=16)

for i, feature in enumerate(category_features):
    color = colors[i % len(colors)]
    counts = X[feature].value_counts()
    axes[i].bar(counts.index.astype(str), counts.values, color=color, edgecolor="black")
    axes[i].tick_params(axis='x', rotation=-60)
    axes[i].set_title(feature)
    if i == 0:
        axes[i].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
binary_features = student_performance.variables[(student_performance.variables['type'] == 'Binary') & (student_performance.variables['role'] == 'Feature')]['name'].tolist()
fig, axes = plt.subplots(nrows=1, ncols=len(binary_features), figsize=(15, 5))

fig.suptitle('Distribution of Binary Features', fontsize=16)

for i, feature in enumerate(binary_features):
    color = colors[i % len(colors)]
    counts = X[feature].value_counts()
    axes[i].bar(counts.index.astype(str), counts.values, color=color, edgecolor="black")
    axes[i].set_title(feature)
    if i == 0:
        axes[i].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
